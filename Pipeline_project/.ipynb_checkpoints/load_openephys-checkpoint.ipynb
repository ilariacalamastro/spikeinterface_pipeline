```python
import spikeinterface.full as si
from pathlib import Path
```

```python
# OPENEPHYS-SPECIFIC PARAMETERS
stream_name  = 'Signals CH'
# run this to list available streams:
# print(si.get_neo_streams('openephys', openephys_folder))

freq_min       = 300.    # highpass cutoff (Hz)
cref_operator  = 'median'
cref_reference = 'global'
```

```python
# LOAD
raw_rec = si.read_openephys(openephys_folder, stream_name=stream_name, load_sync_channel=False)
print(raw_rec)
```

```python
# PREPROCESSING
rec1 = si.highpass_filter(raw_rec, freq_min=freq_min)
bad_channel_ids, _ = si.detect_bad_channels(rec1)
rec2 = rec1.remove_channels(bad_channel_ids)
rec  = si.common_reference(rec2, operator=cref_operator, reference=cref_reference)
print(rec)
```

```python
# SPIKE SORTING 
job_kwargs = dict(n_jobs=4, chunk_duration='1s', progress_bar=True)
rec = rec.save(folder=base_folder / 'preprocess', format='binary', overwrite=True, **job_kwargs)

sorting = si.run_sorter(
    'kilosort4', rec,
    folder=base_folder / 'kilosort_output',
    remove_existing_folder=True, verbose=True)
print(sorting)
```

```python
# QUICK RESULTS OVERVIEW
import matplotlib.pyplot as plt

unit_ids = sorting.get_unit_ids()
duration = sorting.get_total_duration()

print(f"Units found: {len(unit_ids)}")
for uid in unit_ids:
    spikes = sorting.get_unit_spike_train(uid, segment_index=0)
    print(f"  Unit {uid:>4}: {len(spikes):>5} spikes  |  {len(spikes)/duration:.1f} Hz")

# ── AUTOCORRELOGRAMS (no extra compute) ───────────────────────
si.plot_autocorrelograms(sorting)
plt.suptitle("Autocorrelograms")
plt.tight_layout()
plt.show()

# ── SORTING ANALYZER (waveform shapes + amplitudes) ───────────
analyzer = si.create_sorting_analyzer(sorting=sorting, recording=rec, format="memory")
analyzer.compute(["random_spikes", "waveforms", "templates", "noise_levels", "spike_amplitudes"])

si.plot_unit_templates(analyzer)
plt.suptitle("Unit templates")
plt.tight_layout()
plt.show()

si.plot_amplitudes(analyzer)
plt.suptitle("Amplitudes over time")
plt.tight_layout()
plt.show()
```
