```python
import spikeinterface.full as si
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

#YOUR PATHS 
spikeglx_folder = Path(r'C:\Users\labuser\Ilaria\Project\small_rec\2142_g0\2142_g0_imec0')
base_folder     = Path(r'C:\Users\labuser\Ilaria\Project\processed\2142_output')

# check streams
stream_names, stream_ids = si.get_neo_streams(
    'spikeglx', spikeglx_folder)
print(stream_names)

# load
raw_rec = si.read_spikeglx(spikeglx_folder,
    stream_name='imec0.ap', load_sync_channel=False)
print(raw_rec)

# preprocessing 
rec1 = si.highpass_filter(raw_rec,
           freq_min=300.)             # ← CHANGE 2: 400→300 Hz
                                       
bad_channel_ids, _ = si.detect_bad_channels(rec1)
rec2 = rec1.remove_channels(bad_channel_ids)
#rec3 = si.phase_shift(rec2)
rec4 = si.common_reference(rec2,
           operator="median", reference="global")
rec = rec4

# save binary — keep, but change n_jobs and folder
job_kwargs = dict(
    n_jobs=4,                         # ← CHANGE 3: 40→4 threads
    chunk_duration='1s',
    progress_bar=True)
rec = rec.save(
    folder=base_folder / 'preprocess 6', # ← uses your base_folder
    format='binary',
    **job_kwargs)

# run sorter 
sorting = si.run_sorter(
    'kilosort4', rec,
    folder=base_folder / 'kilosort_output', # ← CHANGE 4
    remove_existing_folder=True, verbose=True)
print(sorting)
```

```python
# ── Step 1: Install (run once in terminal) ───────────────────
# pip install unitrefine

# ── Step 2: Import ───────────────────────────────────────────
import spikeinterface.curation as sc
import spikeinterface.widgets as sw

# ── Step 3: Apply UnitRefine ─────────────────────────────────
# Source: official SI auto_label_units tutorial
# analyzer = your SortingAnalyzer from the metrics step
# (the same object you called analyzer.compute([...]) on)

unitrefine_labels = sc.unitrefine_label_units(
    sorting_analyzer=analyzer,

    # Two pre-trained models from HuggingFace Hub:
    # Step 1: noise vs neural classifier
    noise_neural_classifier="SpikeInterface/UnitRefine_noise_neural_classifier",

    # Step 2: SUA vs MUA classifier (applied to neural units only)
    sua_mua_classifier="SpikeInterface/UnitRefine_sua_mua_classifier",
)

# ── Step 4: Inspect the labels ───────────────────────────────
# unitrefine_labels is a pandas DataFrame
print(unitrefine_labels["unitrefine_label"].value_counts())
# Expected output (from tutorial):
# unitrefine_label
# sua      62
# noise    47
# mua      33
# Name: count, dtype: int64

# ── Step 5: Visualise the labelled units ─────────────────────
# Also from the tutorial — plots waveforms coloured by label
w = sw.plot_unit_labels(
    analyzer,
    unitrefine_labels["unitrefine_label"],
    ylims=(-300, 100)
)
w.figure.suptitle("UnitRefine labeling")

# ── Step 6: Save labels to CSV ───────────────────────────────
unitrefine_labels.to_csv(base_folder / 'curation_labels.csv')

# ── Step 7: Get only the good units ──────────────────────────
good_unit_ids = unitrefine_labels[
    unitrefine_labels["unitrefine_label"] == "sua"
].index.tolist()

print(f"Good (SUA) units: {len(good_unit_ids)}")
print(f"Total units:      {len(unitrefine_labels)}")
```
